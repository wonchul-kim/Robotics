```python
import numpy as np
import matplotlib.pyplot as plt
from cs import CanonicalSystem
```

```python
# define system parameters
dt = 0.001
runTime = 2.0
pattern = 'discrete'

cs = CanonicalSystem(dt=dt, runTime=runTime, pattern=pattern)
```

```python
# normal state track w/o error
x_traj1 = np.zeros(cs.n_step)
for i in range(cs.n_step):
    x_traj1[i] = cs.step()

# state track w/ error
cs.reset()
x_traj2 = np.zeros(cs.n_step)
error = np.zeros(cs.n_step)
error[200:400] = 1
for j in range(cs.n_step):
    x_traj2[j] = cs.step(error=error[j])
```

```python
fig, ax1 = plt.subplots(figsize=(6,3))
ax1.plot(x_traj1, lw=2)
ax1.plot(x_traj2, lw=2)
plt.grid()
plt.legend(['w/o error', 'w/ error'])
ax2 = ax1.twinx()
ax2.plot(error, 'r-', lw=2)
plt.legend(['error'], loc='lower right')
plt.ylim(0, 3.5)
plt.xlabel('time (s)')
plt.ylabel('x')
plt.title('Canonical system - discrete')


for t1 in ax2.get_yticklabels():
    t1.set_color('r')

plt.tight_layout()

plt.show()
```

```python
dt = 0.001
runTime = 2*np.pi
pattern = 'rhythmic'

cs = CanonicalSystem(dt=.001, runTime=runTime, pattern=pattern)
x_traj3 = np.zeros(cs.n_step)
for k in range(cs.n_step):
    x_traj3[k] = cs.step()

import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(figsize=(6,3))
ax1.plot(x_traj3, lw=2)
plt.grid()
plt.legend(['normal rollout'], loc='lower right')
plt.xlabel('time (s)')
plt.ylabel('x')
plt.title('Canonical system - rhythmic')
plt.show()
```
